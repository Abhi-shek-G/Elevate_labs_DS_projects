```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("shark_tank_india_cleaned.csv")
df.head()
```

```python
df.info()
df.describe()
```

### **Industry Investment Analysis**

```python
industry_funding = df.groupby("Industry")["Amount_Invested"].sum().sort_values(ascending=False)
print(industry_funding)
```

**Visualization**

```python
industry_funding.plot(kind="bar")
plt.title("Total Investment by Industry")
plt.ylabel("Investment Amount")
plt.show()
```

### **Deal Success Rate**

```python
df["Deal_Success"].value_counts()
```

 **Visualization**

```python
sns.countplot(data=df, x="Deal_Success")
plt.title("Deal Success Distribution")
plt.show()
```

### **Top Investors**

```python
df["Investors"].value_counts().head(10)
```

### **Funding Negotiation**

```python
plt.scatter(df["Amount_Asked"], df["Amount_Invested"])
plt.xlabel("Amount Asked")
plt.ylabel("Amount Invested")
plt.title("Funding Negotiation")
plt.show()
```
